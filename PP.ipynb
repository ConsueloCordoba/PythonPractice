```python
citynumbers=first.groupby('Location')['Degree'].count().reset_index()
```

```python
citynumbers
```

```python
citynumbers.sort_values('Location', ascending=False)
```

```python
cities=first.groupby('Location')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False) \
                             .head(10)
```

```python
cities
```

```python
industry=first.groupby('Industry')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False) \
                             .head(10)
```
