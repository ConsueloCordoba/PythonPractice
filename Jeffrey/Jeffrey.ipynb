```python
df = pd.read_csv('dux-soup-visit-data@2021-04-15 09h45.csv')
```

```python
First=df['Degree']==1
```

```python
First
```

```python
first=df[First]
```

```python
first
```

```python
cities=first.groupby('Location')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False)
```

```python
cities
```

```python
cities.to_csv('cities.csv')
```

```python
industry=first.groupby('Industry')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False)
```

```python
industry
```

```python
industry.to_csv('industry.csv')
```

```python
title=first.groupby('Title')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False)
```

```python
title
```

```python
title.to_csv('title.csv')
```

```python
company=first.groupby('Company')['Degree'] \
                             .count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False)
```
